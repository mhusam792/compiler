```python
def blur_image(image_path):
    """
      A function named 'blur_image' that reads an image from the specified file path, applies a blur filter, and saves the resulting image.

      Parameters:
      - image_path: The path to the input image file.
    """
    base_name=os.path.basename(image_path)
    (file_Name,ext)=os.path.splitext(base_name)
    image = cv2.imread(image_path)
    blurred_image = cv2.blur(image, (61,61))
    cv2.imwrite(f"blured_image/{file_Name} full_blured_image.png",blurred_image)

# blur_image("/content/test2.jpeg")
```

```python
def blur_card2(image_path,card_type):
  """
    A function named 'blur_card2' that reads an image, processes it based on the specified 'card_type', applies a Gaussian blur to specific regions,and saves the resulting image.

    Parameters:
    - image_path: The path to the input image file.
    - card_type: A string specifying the type of card to process ('t' for type Tamween, 'q' for type Egyptian national ID number, 'r' for type Egyptian license).
  """
  base_name=os.path.basename(image_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image = cv2.imread(image_path)
  processed_image = cv2.resize(image, (700, 480))
  if card_type=="t":
    id = processed_image[350:400 ,  350:685]
    id = cv2.GaussianBlur(id, (81,81), 0)
    processed_image[350:400 ,  350:685]=id
    id_2 = processed_image[390:438 , 398:680]
    id_2 = cv2.GaussianBlur(id_2, (81,81), 0)
    processed_image[390:438 , 398:680]=id_2
    cv2.imwrite(f"blured_image/{file_Name} blured_image.png",processed_image)
  elif card_type=='q':
    id = processed_image[360:430 , 300:]
    id = cv2.GaussianBlur(id, (81,81), 0)
    processed_image[360:430 , 300:]=id
    l_name = processed_image[165:220, 340:]
    l_name = cv2.GaussianBlur(l_name, (81,81), 0)
    processed_image[165:220, 340:]=l_name
    cv2.imwrite(f"blured_image/{file_Name} blured_image.png",processed_image)
  elif card_type=='r':
    id = processed_image[65:115 , 200:470]
    id = cv2.GaussianBlur(id, (81,81), 0)
    processed_image[65:115 , 200:470]=id
    name = processed_image[105:180, 200:570]
    name = cv2.GaussianBlur(name, (81,81), 0)
    processed_image[105:180, 200:570]=name
    cv2.imwrite(f"blured_image/{file_Name} blured_image.png",processed_image)

# blur_card2("/content/tamween2.png","t")
```

```python
app = FastAPI()


@app.get("/get_tamweencard_data",tags=["Card"])
async def tamween_card_data(file_path):
    return tamweencard.get_final_data(file_path)

@app.get("/get_plate_num",tags=["Read Data"])
async def get_plate(file_path):
    return egy_plate_recognition.get_Letters_and_nums(file_path)

@app.get("/apply_median_filter",tags=["Image optimization"])
async def median_filter(file_path):
  base_name=os.path.basename(file_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image_enhacment.apply_median_filter(file_path)
  out_put_path=f"/content/outputs/median_filter/{file_Name}.png"
  return {'output_img_path':out_put_path}

@app.get("/get_card",tags=["Card"])
async def get_Card(file_path):
  base_name=os.path.basename(file_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image_enhacment.get_card(file_path)
  out_put_path=f"/content/outputs/cards/{file_Name}.png"
  return {'output_img_path':out_put_path}

@app.get("/resize",tags=["Image optimization"])
async def Resize(file_path,width:int,height:int):
   base_name=os.path.basename(file_path)
   (file_Name,ext)=os.path.splitext(base_name)
   image_enhacment.resize(file_path,width,height)
   out_put_path=f"/content/outputs/resized_image/{file_Name}.png"
   return {'output_img_path':out_put_path}

@app.get("/rotate",tags=["Image optimization"])
async def Rotate(file_path,scale:float,direction:str,angle_direction:str,angle:int):
  base_name=os.path.basename(file_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image_enhacment.rotate(file_path,scale,direction,angle_direction,angle)
  out_put_path=f"/content/outputs/rotated_image/{file_Name}.png"
  return {'output_img_path':out_put_path}

@app.get("/remove_noise",tags=["Image optimization"])
async def remove_Noise(file_path,method):
  base_name=os.path.basename(file_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image_enhacment.remove_noise(file_path,method)#avg_blur gauss_blur bilateral
  out_put_path=f"/content/outputs/DeNoisesImage/{method}/{file_Name}.png"
  return {'output_img_path':out_put_path}

@app.get("/bright_an_img",tags=["Image optimization"])
async def bright_An_Img(file_path):
  base_name=os.path.basename(file_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image_enhacment.bright_an_img(file_path)
  out_put_path=f"/content/outputs/brightness images/{file_Name}.png"
  return {'output_img_path':out_put_path}

@app.get("/light_image",tags=["Image optimization"])
async def light_Image(file_path):
  base_name=os.path.basename(file_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image_enhacment.light_image(file_path)
  out_put_path=f"/content/outputs/light_images/{file_Name}.png"
  return {'output_img_path':out_put_path}

@app.get("/sharpe_image",tags=["Image optimization"])
async def sharpe_Image(file_path):
  base_name=os.path.basename(file_path)
  (file_Name,ext)=os.path.splitext(base_name)
  image_enhacment.sharpe_image(file_path)
  out_put_path=f"/content/outputs/sharpe_images/{file_Name}.png"
  return {'output_img_path':out_put_path}

@app.get("/get_data",tags=["Read Data"])
async def get_data(file_path):
  data=read_any_image.data_corection(read_any_image.get_data(file_path))
  return {"data":data}

@app.get("/create_img")
async def image_genarate(method,text,img_number=6):
  path=''
  keyword=text_to_image_genarate.create_img(method,text,int(img_number))
  if method== 'google':
    path=f'/content/simple_images/{keyword}'
  elif method== 'AI':
    path=f"/content/genarated_images/{keyword}.png"
  return {'output_img_path':path}

@app.get("/blur_image_card",tags=["Card"])
async def blur_image(file_path,type_of_card,card):
  if card =="yes":
    blur_card2(file_path,type_of_card)
    base_name=os.path.basename(file_path)
    (file_Name,ext)=os.path.splitext(base_name)
    out_put_path=f"blured_image/{file_Name} blured_image.png"
  elif card =="no":
    blur_image(file_path)
    base_name=os.path.basename(file_path)
    (file_Name,ext)=os.path.splitext(base_name)
    out_put_path=f"blured_image/{file_Name} full_blured_image.png"
  return {'output_img_path':out_put_path}

```
